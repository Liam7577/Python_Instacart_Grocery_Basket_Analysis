```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'final_merged.pkl'))
```

```python
## Create a graph to show the value counts of the price range loc to help show how many items are in each group

bar_price_range_loc = df['price_range_loc'].value_counts().plot.bar(color =['purple','green', 'blue'])
```

```python
## export graph

bar_price_range_loc.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_price_range_loc.png'))
```

```python
## create a value counts for the different departments to see which ones are the most popular.

bar_departments = df['department'].value_counts().plot.bar()
```

```python
bar_departments.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_departments.png'))
```
