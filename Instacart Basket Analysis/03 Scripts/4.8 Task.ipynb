```python
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Prepared Data', '4.8_reading_ords_prods.pkl'))
```

```python
## Group by department id and return the mean of the order number

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# 5. 
## Group by loyalty flag and and return descriptive statistics on prices of products

ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

```python
# 6.
## Create a spending flag for each user based on average price

ords_prods_merge['mean_spending_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Create a spending flag column that assigns high and low spender based on average spending price of customers orders

ords_prods_merge.loc[ords_prods_merge['mean_spending_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spending_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
## Check value count and relevent columns

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['mean_spending_price','user_id','spending_flag']].head(20)
```

```python
# 7.
## Create a frequency flag for each customer

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Create a flag that assigns order frequency flag

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
## Check value count and relevent columns

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
