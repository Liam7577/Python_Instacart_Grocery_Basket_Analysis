```python
# 3. First lets sample the data as the data set is very large

np.random.seed(3)
dev = np.random.rand(len(df_all)) <= .7
```

```python
## divide the data set up into the 30% and 70%

big = df_all[dev]
small = df_all[~dev]
```

```python
## Now that there are samples we are going to use the smaller sample to create a histogram

hist = df_all['order_hour_of_day'].plot.hist(bins = 24)
```

```python
## This histogram shows the frequency of orders at certain hours of the day. From 10-15 o'clock seems to be where there are the most orders.
```

```python
# 4.
## Create bar chart based off the loyalty flag

bar = df_all['loyalty_flag'].value_counts().plot.bar()
```

```python
# 5. 
## Create line chart

line = sns.lineplot(data = small, x = 'order_hour_of_day',y = 'prices')
```

```python
# 6.
## Create another line graph with age and number of dependents

line_2 = sns.lineplot(data = small, x = 'Age', y = 'n_dependants')
```

```python
## This linechart does not show any distinct pattern. This means that as age goes up there is no direct correlation with more or less dependents
```

```python
# 7.
## Create scatterplot for age and income

scatter = sns.scatterplot(data = small, x = 'Age', y = 'income')
```

```python
## This scatterplot shows that there once people hit the age of 40 there are more outliers of higher earners. However this scatterplot also shows that there is a wide range on incomes and that the majority of people are under 20000 regardless of age.
```
